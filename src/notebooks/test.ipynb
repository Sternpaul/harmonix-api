```python
import requests
import csv

# Function to get customers from a specific page
def get_customers(page, api_key):
    url = f"https://apisandbox.spend.nium.com/api/v1/client/1e1f84b7-3181-4969-bbc2-e3e27f3c0799/customers"
    headers = {
        'x-api-key': api_key
    }
    params = {
        'order': 'desc',
        'page': page,
        'size': 100
    }
    
    response = requests.get(url, headers=headers, params=params)
    return response.json() if response.status_code == 200 else []

# Function to get cards for a specific customer
def get_cards(customer_hash_id, wallet_hash_id, api_key):
    url = f"https://apisandbox.spend.nium.com/api/v1/client/1e1f84b7-3181-4969-bbc2-e3e27f3c0799/customer/{customer_hash_id}/wallet/{wallet_hash_id}/cards"
    headers = {
        'x-api-key': api_key
    }
    
    response = requests.get(url, headers=headers)
    return response.json() if response.status_code == 200 else {}

# Function to save customers with cards to a CSV file
def save_customers_with_cards_to_csv(api_key, csv_file_path):
    with open(csv_file_path, mode='w', newline='') as csv_file:
        fieldnames = ['customer_hash_id', 'wallet_hash_id']
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writeheader()
        
        # Iterate through pages 0 to 10
        for page in range(8, 15):
            customers = get_customers(page, api_key)
            
            for customer in customers:
                customer_hash_id = customer.get('customerHashId')
                wallet_hash_id = customer.get('walletHashId')
                
                # Check if customer has any cards
                cards = get_cards(customer_hash_id, wallet_hash_id, api_key)
                if cards.get('content'):
                    writer.writerow({'customer_hash_id': customer_hash_id, 'wallet_hash_id': wallet_hash_id})

# Replace 'your_api_key_here' with the actual API key and specify the desired CSV file path
api_key = 'L5RHYt6J4qaGI6J4ycOZz1invgQkcwNi643q6S3D'
csv_file_path = 'customers_with_cards_1.csv'

# Call the function to save customers with cards to CSV
save_customers_with_cards_to_csv(api_key, csv_file_path)
```
